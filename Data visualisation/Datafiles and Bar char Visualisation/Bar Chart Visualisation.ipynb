```python
# import libraries
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
sns.set()
```

```python
#  Load the data set
df_survey = pd.read_csv(r'Motors.csv')
```

```python
df_survey
```

```python
plt.figure(figsize = (9,6))
plt.bar( x= df_survey["Automaker"],
           height = df_survey["Numbers"],
        color = "midnightblue")
plt.xticks(rotation = 45, fontsize = 13)
plt.yticks(fontsize = 13)
plt.title("Numbers of Car by Brand", fontsize = 16, fontweight = "bold")
plt.ylabel("Number of Automakers", fontsize = 13)
plt.show()
plt.savefig("Used Cars Bar.png")
```

```python
sns.set_palette('colorblind')
```

```python
# Pie Char Visualisation
plt.figure(figsize = (10,8))
plt.pie(df_survey['Numbers'],
        labels= df_survey['Automaker'].values,
        autopct= '%.2f%%',
        textprops= {'size' : 'x-large',
                    'fontweight' : 'bold',
                    'rotation' : '30',
                    'color' : 'w'})
plt.legend(loc="center")
plt.title('Numbers of Car by Brand', fontsize = 18, fontweight = 'bold', color = 'pink')
plt.show()
plt.savefig("Used Cars Bar.png")
```

```python
# Stack plot Visualisation
colors = ['#011638', "#7e2987"]
labels = ['Automaker', 'Numbers']
sns.set_style("white")
plt.figure(figsize = (9,6))
plt.stackplot(df_survey['Automaker'],
              df_survey['Numbers'],
              colors = colors,
              edgecolor = 'none')
plt.xticks(rotation = 45)
plt.legend(labels = labels, loc = 'upper left')
plt.ylabel('Numbers of Car by Brand', fontsize = 13)
plt.title("Number of Automakers")
sns.despine()
plt.show()
```

```python
# colors = ['#011638', "#7e2987"]
# labels = ['Automaker', 'Numbers']
sns.set_style("white")
df_survey.plot(kind='bar', stacked=True,figsize=(10, 8), color = 'midnightblue')
plt.legend(labels = labels, loc = 'upper left')
plt.xticks(rotation = 45)
plt.title("Number of Automakers")
plt.xlabel("Automaker")
plt.ylabel("Numbers")
sns.despine()
plt.show()
```

```python
# Visualisation on Bar Chart
plt.figure(figsize = (9,6))
plt.barh( width= df_survey["Numbers"],
           y = df_survey["Automaker"],
        color = "midnightblue")
plt.xticks(rotation = 45, fontsize = 13)
plt.yticks(fontsize = 13)
plt.title("Car Listing by Brand", fontsize = 16, fontweight = "bold")
plt.ylabel("Number of Listings", fontsize = 13)
plt.show()
plt.savefig("Used Cars Bar.png")
```

```python
CFA = pd.read_csv('lending-co-LT.csv')
```

```python
CFA
```

```python
import seaborn as sns
sns.set()
plt.figure(figsize = (10,8))
sns.scatterplot(x =CFA['LoanID'],
                y = CFA['TotalPrice'],
                palette = ['black', 'darkblue', 'purple', 'pink', 'white'],
                s = 100)
plt.title("Relationship between LoanID and TotalPrice plot",
          fontsize = 14,
          weight = 'bold')
plt.xlabel("LoanID", weight = 'bold')
plt.ylabel("TotalPrice", weight = 'bold')
plt.show()
```

```python
# Linear Regression plot Visualisation
plt.figure(figsize=(10,8))
sns.set(rc = {'figure.figsize':(9,4)})
sns.regplot(x = "LoanID",
            y = "TotalPrice",
            data = CFA,
            scatter_kws= {'color': 'blue'},
            line_kws= {'color' : 'blue'})
plt.xlabel("LoanID")
plt.ylabel("TotalPrice")
plt.title("Scatter plot using line", fontsize = 14, weight = 'bold')
plt.show()
```
